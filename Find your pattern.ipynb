```python
import os
import numpy as np
from PIL import Image

from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
import time
from selenium.webdriver.chrome.options import Options
from options import profile_path

import matplotlib.pyplot as plt
```

```python
# Create the options
options = webdriver.ChromeOptions()
# Add your chrome profile (to remember you) and specify full screen
options.add_argument("user-data-dir={}".format(profile_path)) #Path to your chrome profile
options.add_argument("--start-maximized")
options.add_argument('log-level=3')
```

### Storage of a few screenshot

```python
# Run the driver
driver = webdriver.Chrome(options=options)
# Open instagram
driver.get('https://www.instagram.com/?hl=fr')
time.sleep(3)

for i in range(5) :
    driver.get_screenshot_as_file("./screenshots/screenshot{}.png".format(i))
    driver.execute_script("window.scrollBy(0, 400)")
    time.sleep(.2)
    
driver.quit()
```

```python
im = np.array(Image.open("./screenshots/pattern.png"))[:,:,:3]
current = im
plt.imshow(current)
plt.show()
done = "false"
while done == "false" :
    x1,x2,y1,y2 = list(map(int,input("x1, x2, y1, y2 = ? ").split(" ")))
    assert x1 > 0 and x2 > x1 and y1 > 0 and y2 > 0
    current = im[x1:x2,y1:y2,:]
    plt.imshow(current)
    plt.show()
    done = input("Finished ? Awnser by true or false. ")

Image.fromarray(current).save("./data/pattern.png")
print("You need to copy these values in options.py")
print(f"y1 = {y1}, y2 = {y2}")
```
